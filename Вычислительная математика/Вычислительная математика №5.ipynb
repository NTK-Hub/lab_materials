**Задание 1** \\
Найти точное и приближенные значения первой производной функции индивидуального задания в точке x (конкретное, не обращающее функцию в 0 значение, выбрать самостоятельно), использовав формулы численного дифференцирования первого, второго и четвертого порядка аппроксимации, вычислив или подобрав оптимальный шаг h. Для каждого полученного значения определить абсолютную, относительную погрешности (сопоставив точное и приближенные значения), проверить теоретическую оценку абсолютной погрешности r.

\begin{equation*}
f'(x) = 2\cdot e^{0.1x} (-2\cdot sin(2x)) + (-0.2\cdot e^{-0.1x}) \cdot cos(2x) - 0.3\cdot cos(0.3x)
\end{equation*}

При x = 0

f'(0) = -0.5

```python
import math

def f(x):
  return 2 * math.exp(-0.1 * x) * math.cos(2 * x) - math.sin(0.3 * x)

def first(x):
  h = 0.0001

  return (f(x + h) - f(x)) / h

actual = -0.5
print(first(0))
print("Абсолютная погрешность: ", abs(actual - first(0)))
print("Относительная погрешность: ", abs(actual - first(0)) / actual)
```

**Задание 2** \\
Найти точное и приближенное значения второй производной функции индивидуального задания в точке x. Подобрать оптимальный шаг h экспериментально. Определить абсолютную, относительную погрешности.

\begin{equation*}
f''(x) = -0.4 \cdot e^{-0.1x} sin(2x) + (0.4 \cdot e^{-0.1x}) \cdot cos(2x) + 0.3 \cdot sin(0.3x)
\end{equation*}

f'(0) = 0.4

```python
import math

def f(x):
  return 2 * math.exp(-0.1 * x) * math.cos(2 * x) - math.sin(0.3 * x)

def differ2(x):
  h = 69

  return (f(x - h) - 2 * f(x) + f(x + h)) / (h ** 2)

actual = 0.4
print(differ2(0))
print("Абсолютная погрешность: ", abs(actual - differ2(0)))
print("Относительная погрешность: ", abs(actual - differ2(0)) / actual)
```

**Задание 3** \\
Найти точное и приближенное значения третьей производной функции индивидуального задания в точке x. Подобрать оптимальный шаг h экспериментально. Определить абсолютную, относительную погрешности.

\begin{equation*}
f'''(x) = 0.04 \cdot e^{-0.1x} sin(2x) - (0.84 \cdot e^{-0.1x}) \cdot cos(2x) - 0.03 \cdot cos(0.3x)
\end{equation*}

f'''(0) = -0.75

```python
import math

def f(x):
  return 2 * math.exp(-0.1 * x) * math.cos(2 * x) - math.sin(0.3 * x)

def differ3(x):
  h = 81.62

  return (f(x + 3*h / 2) - 3 * f(x + h / 2) + 3 * f(x - h / 2) - f(x - 3 * h / 2)) / (h ** 3)

actual = -0.75
print(differ3(0))
print("Абсолютная погрешность: ", abs(actual - differ3(0)))
print("Относительная погрешность: ", abs(actual - differ3(0)) / actual)
```

Подготовить программы на языке Rust для всех численных расчетов.

```python
%%writefile lab5.rs

fn f(x: f64) -> f64 {
    2. * f64::exp(-0.1 * x) * f64::cos(2. * x) - f64::sin(0.3 * x)
}

fn first(x: f64) -> f64 {
    let h = 0.0001;
    return (f(x + h) - f(x)) / h;
}

fn second(x: f64) -> f64 {
    let h = 0.02;
    return (f(x + h) - f(x - h)) / (2. * h);
}

fn fourth(x: f64) -> f64 {
    let h = 0.12;
    return (f(x - 2. * h) - 8. * f(x - h) + 8. * f(x + h) - f(x + 2. * h)) / (12. * h);
}

fn differ2(x: f64) -> f64 {
    let h = 69.;
    return (f(x - h) - 2. * f(x) + f(x + h)) / f64::powi(h, 2);
}

fn differ3(x: f64) -> f64 {
    let h = 81.62;
    return (f(x + 3.*h / 2.) - 3. * f(x + h / 2.) + 3. * f(x - h / 2.) - f(x - 3. * h / 2.)) / f64::powi(h, 3);
}

fn main() {
    let x = 0.;

    println!("Первая производная первого порядка: \n");
    let actual = -0.5;
    println!("Абсолютная погрешность: {}", actual - first(x));
    println!("Относительная погрешность: {}\n\n", (actual - first(x)) / actual);

    println!("Первая производная второго порядка: \n");
    let actual = -0.5;
    println!("Абсолютная погрешность: {}", actual - second(x));
    println!("Относительная погрешность: {}\n\n", (actual - second(x)) / actual);

    println!("Первая производная четвёртого порядка: \n");
    let actual = -0.5;
    println!("Абсолютная погрешность: {}", actual - fourth(x));
    println!("Относительная погрешность: {}\n\n", (actual - fourth(x)) / actual);

    println!("Вторая производная: \n");
    let actual = 0.4;
    println!("Абсолютная погрешность: {}", actual - differ2(x));
    println!("Относительная погрешность: {}\n\n", (actual - differ2(x)) / actual);

    println!("Третья производная: \n");
    let actual = -0.75;
    println!("Абсолютная погрешность: {}", actual - differ3(x));
    println!("Относительная погрешность: {}\n\n", (actual - differ3(x)) / actual);
}
```
